```python
from tkinter import *
from tkinter import ttk
import serial
import matplotlib.pyplot as plt
ser = serial.Serial('COM3',  baudrate=115200, timeout=1)
```

```python
root = Tk()
root.geometry('820x820')
root['bg']="#333"
ca = Canvas(root, width=800, height=800)
ca.place(x=8,y=8)

dd = [0]*100
outdata = [0]*99

def tt():
    global dd
    global outdata
    ca.delete("all")
    
    for i in range(99):
        dd[i] = dd[i+1]
    dd[99] = int(ser.readline())
    
    
    for i in range(99):
        outdata[i] = 400 - ( dd[i] - dd[i+1] )
        if outdata[i]>300 and outdata[i]<500:
            outdata[i] = ( outdata[i] - 400 ) * 0.3 + 400
           
    print( outdata[98] )
    
    
    xy=[]
    for i in range(len(outdata)):
        xy += ( 800-i*8 , outdata[i] )           
    ca.create_line(xy, width=2, fill='blue')    
    
      
    ca.after(1, tt)

ca.after(1, tt)

mainloop()
```
